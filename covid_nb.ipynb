```python
covid_totals = covid_data.groupby('Country').max()
```

```python
fig, ax = plt.subplots(figsize=(50, 25))
sns.scatterplot(covid_totals['Confirmed'], covid_totals['Death'], s=200)
plt.xticks(size=25)
plt.yticks(size=25)
plt.xlabel('Confirmed', size=25)
plt.ylabel('Deaths', size=25)
plt.show()
```
